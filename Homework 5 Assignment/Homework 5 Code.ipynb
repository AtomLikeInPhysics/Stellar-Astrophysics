```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
bins = np.arange(0.1,21,1)
# print(bins)

a = 1.85e4      #constant calculated from part a

def IMF(m):     #IMF function
    return(a * m**-2.35)


num_stars = []


#Integrate over each bin
for i in range(len(bins)-1):
    num_stars.append(quad(IMF,bins[i], bins[i+1])[0])
print(num_stars)
num_stars = np.array(num_stars)

#make the bins I populate center themselves over the median value between their limits
cent = (bins[:-1] + bins[1:]) / 2
print(cent)

#ploting the histogram
plt.bar(cent,(num_stars),width=1,edgecolor = 'k')
plt.yscale('log')
plt.xlabel('Stellar Mass (M_sun)')
plt.ylabel('Number of Stars Log(N)')
plt.title('IMF for Masses 0.1-20 M_sun')
plt.savefig(fname = 'IMF')
```

```python
def luminosity(m):
    return((1.5 * m**3.5)*IMF(m))

num_stars = []
for i in range(len(bins)-1):
    num_stars.append(quad(luminosity,bins[i], bins[i+1])[0])
print(num_stars)
num_stars = np.array(num_stars)

cent = (bins[:-1] + bins[1:]) / 2

plt.bar(cent,(num_stars),width=1,edgecolor = 'k')
plt.yscale('log')
plt.xlabel('Stellar Mass (M_sun)')
plt.ylabel('Luminosity Log(L/L_sun)')
plt.title('Luminosity for Masses 0.1-20 M_sun')
plt.savefig(fname = 'Luminosity')
```
