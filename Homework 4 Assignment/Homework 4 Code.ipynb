a) Using the output of these simulations, create an HR diagram and a
separate graph of the surface luminosity (y-axis) vs. the surface radius
(x-axis) with all 5 stars over plotted on this one diagram. Be sure
to label which star belongs to which path and additionally mark each
starâ€™s initial and final positions in the graph.

```python
half = np.genfromtxt('half_history.data',names = True, skip_header=5)
one = np.genfromtxt('one_history.data',names = True, skip_header=5)
five = np.genfromtxt('five_history.data',names = True, skip_header=5)
fifteen = np.genfromtxt('fifteen_history.data',names = True, skip_header=5)
fifty = np.genfromtxt('fifty_history.data',names = True, skip_header=5)
```

```python
mass_str = ('half','one','five','fifteen','fifty')
mass = (half,one,five,fifteen,fifty)
for i in range(len(mass)):
    plt.plot(mass[i]['log_Teff'],mass[i]['log_L'],label = f'{(mass_str[i])}')
    plt.scatter(mass[i]['log_Teff'][0],mass[i]['log_L'][0],c='r')
    plt.scatter(mass[i]['log_Teff'][-1],mass[i]['log_L'][-1],c='b')
plt.title('HR-Diagram')
plt.ylabel('Log Luminosity[L/M(solar)]')
plt.xlabel('Log $T_{eff}$ [T/K]')
plt.text(4.6,1.5,"Start",c='r')
plt.text(4.6,1,"End",c='b')
plt.legend()
plt.gca().invert_xaxis()
# plt.colorbar()
# c = np.log(mass[i]['star_age'])
```
